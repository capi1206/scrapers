```python
import requests
from bs4 import BeautifulSoup

url = "https://www.doctoralia.co/especialidades-medicas"
headers = {
    "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_7)"
}

resp = requests.get(url, headers=headers)
resp.raise_for_status()  # Raise an error if request fails

soup = BeautifulSoup(resp.text, "html.parser")

# Find all <a> tags with class="text-muted"
links = soup.find_all("a", class_="text-muted")

specialties = []
for link in links:
    specialty = {
        "name": link.get_text(strip=True),
        "url": "https://www.doctoralia.co" + link["href"]
    }
    specialties.append(specialty)

# Print results
for s in specialties:
    print(s)
```

```python
##Now go to each link in the links list and extract the info
docs = []
for sp in specialties:
    if not "ver m" in sp["name"]:
        link = sp["url"]
        resp = requests.get(link, headers=headers)
        resp.raise_for_status()  # Raise an error if request fails

        soup = BeautifulSoup(resp.text, "html.parser")

        # Find all <a> tags with class="text-body"
        links = soup.find_all("a", class_="text-body")

        docs += links

# <a href="https://www.doctoralia.co/guillermo-palacio/pediatra/barranquilla#address-id=16959&amp;is-online-only=false&amp;filters%5Bspecializations%5D%5B%5D=1" data-id="address-context-cta" class="text-body" data-eec-select-item="">
# 					<span data-tracking-id="result-card-name" data-ga-event="click" data-ga-category="NS Doctor Card" data-ga-action="Detail click" data-ga-label="Doctor Name" data-eecommerce-action="product-click" itemprop="name">
# 						Dr. Guillermo Palacio
# 					</span>
# 				</a>
```
